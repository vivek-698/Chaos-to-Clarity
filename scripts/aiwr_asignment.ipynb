```python
def parse_query(infix_tokens):
    
    precedence = {}
    precedence['NOT'] = 3
    precedence['AND'] = 2
    precedence['OR'] = 1
    precedence['and'] = 2
    precedence['or'] = 1
    precedence['('] = 0
    precedence[')'] = 0    

    output = []
    operator_stack = []

    for token in infix_tokens:
        if (token == '('):
            operator_stack.append(token)

        elif (token == ')'):
            operator = operator_stack.pop()
            while operator != '(':
                output.append(operator)
                operator = operator_stack.pop()

        elif (token in precedence):
            if (operator_stack):
                current_operator = operator_stack[-1]
                while (operator_stack and precedence[current_operator] > precedence[token]):
                    output.append(operator_stack.pop())
                    if (operator_stack):
                        current_operator = operator_stack[-1]
            operator_stack.append(token) # add token to stack
        else:
            output.append(token.lower())

    while (operator_stack):
        output.append(operator_stack.pop())

    return output
```

```python
def boolean_query(query, inverted_index):
    query = query.strip()
    query_tokens = query.split()
    boolean_query = parse_query(query_tokens)
        
    result_stack = list()
    for idx, token in enumerate(boolean_query):
        if token not in ["AND", "NOT", "OR","and","or"]:
            result = set(inverted_index[token])
        else:
            if token in ['AND', 'OR',"and","or"]:
                right_operand = result_stack.pop()
                left_operand = result_stack.pop()
                
                if token == 'AND' or token == 'and':
                    operation = set.intersection
                else:
                    operation = set.union
                
                result = operation(left_operand, right_operand)
                
            else:
                operand = result_stack.pop()
                complement_document_ids = inverted_index[boolean_query[idx-1]]
                result = list()
                for word in inverted_index:
                    result.extend([_id for _id in inverted_index[word] if _id not in complement_document_ids])
                result = set(result)
                
        result_stack.append(result)
    
    return result_stack.pop()

        
```

```python

def get_kgrams(word, k):
    word = f'${word.strip()}$'
    kgrams = ngrams(word, k)
    kgrams = list(map(lambda x: ''.join(x), kgrams))
    return kgrams


def create_kgram_index(articles, k=3):
    kgram_index = dict()
    for article in articles:
        for token in article:
            kgrams = get_kgrams(token, k)
            for kgram in kgrams:
                if kgram not in kgram_index:
                    kgram_index[kgram] = set()
                kgram_index[kgram].add(token)
                
    for kgram in kgram_index:
        kgram_index[kgram] = sorted(list(kgram_index[kgram]))
    
    return kgram_index
```

```python
def preprocess(text):
    # Removing stop words
    #stop_words = set(stopwords.words('english'))
    stop_words = nltk.corpus.stopwords.words('english')
    words = nltk.word_tokenize(text)
    words = [word for word in words if word.lower() not in stop_words]

    # Converting all text to lowercase
    words = [word.lower() for word in words]

    # Removing all punctuation
    words = [word for word in words if word not in string.punctuation]

    # Stemming the words
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]

    #print(words)

    return words
```

```python
#Similarity Index
def create_inverted_index(documents):
    inverted_index = {}
    doc_lengths = {}

    for doc_id, document in enumerate(documents):
        words = preprocess(document)

        # Updating the document length
        doc_lengths[doc_id] = len(words)

        # Creating the mapping between words and documents
        for position, word in enumerate(words):
            if word not in inverted_index:
                inverted_index[word] = {}

            if doc_id not in inverted_index[word]:
                inverted_index[word][doc_id] = {
                    'tf': 0,
                    'positions': []
                }

            inverted_index[word][doc_id]['tf'] += 1
            inverted_index[word][doc_id]['positions'].append(position)

    return inverted_index, doc_lengths
```

```python
#SIMILARITY MATRIX
#retrieve documents based on similarity
import math

def retrieve_documents(query, inverted_index, doc_lengths, top_n=10):
    # Preprocess the query
    query_words = preprocess(query)

    # Calculate the query vector
    query_vector = {}
    for word in query_words:
        if word not in query_vector:
            query_vector[word] = 0
        query_vector[word] += 1

    # Calculate the query length
    query_length = math.sqrt(sum([query_vector[word] ** 2 for word in query_vector]))

    # Calculate the scores for each document
    scores = {}
    for word in query_words:
        if word in inverted_index:
            for doc_id in inverted_index[word]:
                tf = inverted_index[word][doc_id]['tf']
                #tf = 0
                
                positions = inverted_index[word][doc_id]['positions']
                doc_length = doc_lengths[doc_id]

                # Calculate the tf-idf score
                tf_idf = tf * math.log(len(text) / len(inverted_index[word]))

                # Calculate the document vector
                doc_vector = {}
                for position in positions:
                    doc_vector[position] = tf_idf / doc_length

                # Calculate the cosine similarity
                dot_product = sum([query_vector[word] * doc_vector.get(position, 0) for position in doc_vector])
                doc_length_norm = math.sqrt(sum([doc_vector[position] ** 2 for position in doc_vector]))
                cosine_similarity = dot_product / (query_length * doc_length_norm)

                # Update the scores
                if doc_id not in scores:
                    scores[doc_id] = 0
                scores[doc_id] += cosine_similarity

    # Rank the documents based on the scores
    ranked_docs = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]

    # Return the top N documents
    results = []
    for doc_id, score in ranked_docs:
        results.append(text[doc_id])
        #results.append(doc_id)
    return results
```

```python
#Relavant text
def retrieve_text_likelihood(query, inverted_index, doc_lengths, k=10):
    query_words = preprocess(query)
    scores = {}

    for word in query_words:
        if word not in inverted_index:
            continue

        idf = math.log(len(doc_lengths) / len(inverted_index[word]))
        for doc_id, info in inverted_index[word].items():
            tf = info['tf']
            doc_length = doc_lengths[doc_id]
            prob = (tf / doc_length) * idf

            if doc_id not in scores:
                scores[doc_id] = 0.0

            scores[doc_id] += math.log(prob)

    top_k_docs = sorted(scores.items(), key=lambda x: x[1], reverse=False)[:k]
    return [(text[doc_id], score) for doc_id, score in top_k_docs]
```

```python
import numpy as np

def edit_distance(s1, s2):
    m, n = len(s1), len(s2)
    dp = np.zeros((m+1, n+1))
    for i in range(m+1):
        dp[i][0] = i
    for j in range(n+1):
        dp[0][j] = j
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])
    return dp[m][n]

def retrieve_similar_words(word, inverted_index, threshold):
    word = word.lower()
    similar_words = []
    for term in inverted_index.keys():
        distance = edit_distance(word, term)
        if distance <= threshold:
            similar_words.extend(inverted_index[term])
            print(term)
    return list(set(similar_words))
```

```python
def search_phrase_query(query, positional_index):
    query_terms = query.split()
    doc_list = []
    
    # Check if all query terms exist in positional_index
    for term in query_terms:
        if term not in positional_index:
            print(f"Query term '{term}' not found in the corpus")
            return []
    
    # Initialize variables for the first query term
    first_term = query_terms[0]
    first_term_postings = positional_index[first_term]
    first_term_doc_ids = list(first_term_postings.keys())
    first_term_pos_lists = list(first_term_postings.values())
    
    # Iterate over the positions of the first term
    for i in range(len(first_term_pos_lists)):
        doc_id = first_term_doc_ids[i]
        pos = first_term_pos_lists[i][0]
        matched = True
        
        # Iterate over the remaining query terms
        for j in range(1, len(query_terms)):
            term = query_terms[j]
            term_postings = positional_index[term]
            
            # Check if current doc_id contains term
            if doc_id not in term_postings:
                matched = False
                break
            
            # Check if term appears after the previous term
            term_pos_list = term_postings[doc_id]
            found = False
            for pos_index in range(len(term_pos_list)):
                if term_pos_list[pos_index] > pos:
                    pos = term_pos_list[pos_index]
                    found = True
                    break
            if not found:
                matched = False
                break
        
        # Add doc_id to the results if all query terms matched
        if matched:
            doc_list.append(doc_id)
            
    return doc_list


'''
This function works by first checking if all query terms exist in the positional_index. It then initializes variables for the first query term and iterates over the positions of the first term. For each position, it checks if the remaining query terms appear in the same document after the position of the first term. If all query terms match, it adds the document ID to the results. '''


```
